```
# Part A: Theoretical Fundamentals (Maximum Likelihood & Decision Space)

## 1. Bias of the Maximum Likelihood Covariance Estimator

Consider a collection of independent and identically distributed random vectors

[
X_1,X_2,\ldots,X_n \in \mathbb{R}^{m}
]

with population mean

[
\mu = E[X_i]
]

and covariance matrix

[
\Sigma = E[(X_i-\mu)(X_i-\mu)^T].
]

The maximum likelihood estimator (MLE) of the covariance matrix is

[
\widehat{\Sigma}_{MLE}
======================

\frac{1}{n}
\sum_{i=1}^{n}
(X_i-\bar X)(X_i-\bar X)^T,
]

where

[
\bar X=\frac{1}{n}\sum_{i=1}^{n}X_i.
]

### Step 1: Expand the Centered Sum

A standard identity gives

[
\sum_{i=1}^{n}(X_i-\bar X)(X_i-\bar X)^T
========================================

## \sum_{i=1}^{n}(X_i-\mu)(X_i-\mu)^T

n(\bar X-\mu)(\bar X-\mu)^T.
]

Taking expectations on both sides,

[
E!\left[
\sum_{i=1}^{n}(X_i-\bar X)(X_i-\bar X)^T
\right]
=======

E!\left[
\sum_{i=1}^{n}(X_i-\mu)(X_i-\mu)^T
\right]
-------

nE[(\bar X-\mu)(\bar X-\mu)^T].
]

### Step 2: Evaluate Each Expectation

Since the observations are i.i.d.,

[
E[(X_i-\mu)(X_i-\mu)^T]=\Sigma.
]

Therefore,

[
E!\left[
\sum_{i=1}^{n}(X_i-\mu)(X_i-\mu)^T
\right]
=======

n\Sigma.
]

The covariance of the sample mean is

[
Var(\bar X)
===========

\frac{\Sigma}{n}.
]

Hence,

[
E[(\bar X-\mu)(\bar X-\mu)^T]
=============================

\frac{\Sigma}{n}.
]

Substituting,

[
E!\left[
\sum_{i=1}^{n}(X_i-\bar X)(X_i-\bar X)^T
\right]
=======

# n\Sigma-\Sigma

(n-1)\Sigma.
]

### Step 3: Obtain the Expectation of the MLE

Multiplying by the factor (1/n),

[
E[\widehat{\Sigma}_{MLE}]
=========================

\frac{1}{n}(n-1)\Sigma.
]

Thus,

[
\boxed{
E[\widehat{\Sigma}_{MLE}]
=========================

\frac{n-1}{n}\Sigma
}
]

which proves that the MLE systematically underestimates the true covariance matrix.

Therefore, the covariance MLE is biased for finite sample sizes.

---

## Bessel's Correction and the Unbiased Covariance Estimator

To eliminate the bias, replace the denominator (n) with (n-1):

[
S
=

\frac{1}{n-1}
\sum_{i=1}^{n}
(X_i-\bar X)(X_i-\bar X)^T.
]

Taking expectation,

[
E[S]
====

\frac{1}{n-1}
E!\left[
\sum_{i=1}^{n}
(X_i-\bar X)(X_i-\bar X)^T
\right].
]

Using the previous result,

[
E[S]
====

\frac{1}{n-1}(n-1)\Sigma.
]

Hence,

[
\boxed{E[S]=\Sigma}
]

showing that (S) is unbiased.

The correction is necessary because one degree of freedom is consumed when the unknown population mean is replaced by the sample mean.

---

# 2. Type I Error, Type II Error and Diagnostic Decision Space

## Type I Error ((\alpha))

A Type I error occurs when the monitoring system declares that damage or abnormal behaviour exists even though the structure is actually healthy.

Mathematically,

[
\alpha=P(\text{Reject }H_0 \mid H_0 \text{ is true}).
]

In structural health monitoring, this corresponds to a false alarm, potentially causing unnecessary inspections, downtime, or shutdowns.

---

## Type II Error ((\beta))

A Type II error occurs when the monitoring system fails to detect an existing fault.

Mathematically,

[
\beta=P(\text{Fail to reject }H_0 \mid H_1 \text{ is true}).
]

In practice, damage may be present while the system incorrectly reports normal operation.

---

## Effect of Choosing an Extremely Small Significance Level

Suppose the engineer selects

[
\alpha = 0.0001.
]

A smaller significance level raises the critical threshold required to reject (H_0).

Consequently, fewer observations fall into the rejection region.

This reduces the probability of false alarms but simultaneously increases the chance of missing genuine faults.

Therefore,

[
\beta \uparrow
]

and the statistical power

[
Power = 1-\beta
]

decreases.

Hence,

[
\boxed{\alpha \downarrow ;\Rightarrow; \beta \uparrow ;\Rightarrow; Power \downarrow}
]

for a fixed sample size and effect magnitude.

---

## Geometric Interpretation

For multivariate monitoring, healthy behaviour is commonly represented by a confidence ellipsoid such as that generated by Hotelling's (T^2) statistic.

Reducing (\alpha) moves the critical boundary outward because a larger quantile is required to trigger an alarm.

As a result:

* The acceptance region becomes larger.
* The confidence ellipsoid expands.
* More observations are classified as healthy.
* Fault detection sensitivity decreases.

Therefore, an ultra-conservative significance level enlarges the geometric volume of the healthy-operation ellipsoid while reducing the ability of the monitoring system to detect subtle structural degradation.
```

```
## Part B: Asymptotic Distribution Using Slutsky's Theorem

Let (\mathbf{X}_1,\mathbf{X}_2,\ldots,\mathbf{X}_n) be independent and identically distributed random vectors with mean vector (\boldsymbol{\mu}) and covariance matrix (\boldsymbol{\Sigma}). The sample mean vector is defined as

[
\widehat{\boldsymbol{\mu}}*n=\frac{1}{n}\sum*{i=1}^{n}\mathbf{X}_i.
]

According to the Multivariate Central Limit Theorem,

[
\sqrt{n}\left(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu}\right)
\xrightarrow{d}
\mathcal{N}\left(\mathbf{0},\boldsymbol{\Sigma}\right).
]

This implies that for sufficiently large (n),

[
\widehat{\boldsymbol{\mu}}_n
\approx
\mathcal{N}\left(\boldsymbol{\mu},\frac{\boldsymbol{\Sigma}}{n}\right).
]

In practice, the population covariance matrix (\boldsymbol{\Sigma}) is usually unknown. Therefore, it is estimated using the sample covariance matrix

[
\mathbf{S}
==========

\frac{1}{n-1}
\sum_{i=1}^{n}
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)^T.
]

Since (\mathbf{S}) is a consistent estimator of (\boldsymbol{\Sigma}),

[
\mathbf{S}\xrightarrow{p}\boldsymbol{\Sigma}.
]

Slutsky's Theorem states that if a sequence of random variables converges in distribution and another sequence converges in probability to a constant, then the limiting distribution remains unchanged when the constant is replaced by its consistent estimator.

Applying Slutsky's Theorem, we may replace (\boldsymbol{\Sigma}) with (\mathbf{S}) in the asymptotic distribution of the sample mean. Therefore,

[
\sqrt{n}\left(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu}\right)
\xrightarrow{d}
\mathcal{N}\left(\mathbf{0},\mathbf{S}\right),
]

and equivalently,

[
\widehat{\boldsymbol{\mu}}_n
\approx
\mathcal{N}\left(\boldsymbol{\mu},\frac{\mathbf{S}}{n}\right).
]

Hence, the asymptotic distribution of the sample mean can be expressed using the estimated covariance matrix, which provides a practical basis for statistical inference and confidence region construction in multivariate analysis.
```
