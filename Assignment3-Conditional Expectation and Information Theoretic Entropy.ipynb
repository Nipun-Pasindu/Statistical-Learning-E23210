## #Q1) Prediction of fraudulent transactions

### 1. Probability Space $(\Omega,\mathfrak F,P)$

The sample space is given by:


$$\Omega = \{\omega_1, \omega_2, \omega_3, \omega_4\}$$

The $\sigma$-algebra $\mathfrak F$ is the power set $\mathfrak P(\Omega)$, containing all $2^4 = 16$ possible subsets:


$$\mathfrak F = \big\{ \emptyset, \{\omega_1\}, \{\omega_2\}, \{\omega_3\}, \{\omega_4\}, \{\omega_1,\omega_2\}, \{\omega_1,\omega_3\}, \{\omega_1,\omega_4\}, \{\omega_2,\omega_3\}, \{\omega_2,\omega_4\}, \{\omega_3,\omega_4\}, \{\omega_1,\omega_2,\omega_3\}, \{\omega_1,\omega_2,\omega_4\}, \{\omega_1,\omega_3,\omega_4\}, \{\omega_2,\omega_3,\omega_4\}, \Omega \big\}$$

The probability measure $P$ on singletons is specified as:


$$P(\{\omega_1\})=0.50, \quad P(\{\omega_2\})=0.10, \quad P(\{\omega_3\})=0.20, \quad P(\{\omega_4\})=0.20$$


For any other set $A \in \mathfrak F$, $P(A) = \sum_{\omega_i \in A} P(\{\omega_i\})$.

### 2. $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$

The random variable $Y$ partitions the sample space into two pre-images:

* $Y^{-1}(0) = \{\omega_1, \omega_2\}$
* $Y^{-1}(1) = \{\omega_3, \omega_4\}$

The $\sigma$-algebra $\sigma(Y)$ consists of all unions, intersections, and complements of these structural atoms:


$$\sigma(Y) = \big\{ \emptyset, \{\omega_1, \omega_2\}, \{\omega_3, \omega_4\}, \Omega \big\}$$

### 3. $\sigma(Y)$-measurability of $E[X\mid Y]$

By definition, the conditional expectation $E[X \mid Y]$ is a random variable that is measurable with respect to the information context available, which is $\sigma(Y)$.

For a random variable to be measurable with respect to a specific $\sigma$-algebra, its value cannot vary across the elements of any indivisible set (atom) within that $\sigma$-algebra. Because the atoms of $\sigma(Y)$ are $\{\omega_1, \omega_2\}$ and $\{\omega_3, \omega_4\}$, $E[X \mid Y]$ cannot distinguish between elements inside the same atom. Hence, it must be constant on $\{\omega_1, \omega_2\}$ and constant on $\{\omega_3, \omega_4\}$.

### 4. Compute $E[X\mid Y]$ explicitly

We calculate the expected value of $X$ given each value of $Y$:

* **For $Y = 0$:**

$$E[X \mid Y=0] = \frac{P(X=1, Y=0)}{P(Y=0)} = \frac{P(\{\omega_2\})}{P(\{\omega_1, \omega_2\})} = \frac{0.10}{0.50 + 0.10} = \frac{0.10}{0.60} = \frac{1}{6}$$


* **For $Y = 1$:**

$$E[X \mid Y=1] = \frac{P(X=1, Y=1)}{P(Y=1)} = \frac{P(\{\omega_4\})}{P(\{\omega_3, \omega_4\})} = \frac{0.20}{0.20 + 0.20} = \frac{0.20}{0.40} = \frac{1}{2}$$



Expressing this operational logic back as an indicator function over $\Omega$:


$$E[X\mid Y]= \frac{1}{6} \mathbb{1}_{\{Y=0\}} + \frac{1}{2} \mathbb{1}_{\{Y=1\}}$$

### 5. Verify the Defining Property $\int_A E[X\mid Y]\,dP=\int_A X\,dP$

We must check this equality for all four sets $A \in \sigma(Y)$:

1. **$A = \emptyset$:** Both sides evaluate identically to $0$.
2. **$A = \{\omega_1, \omega_2\}$ (where $Y=0$):**

$$\int_A E[X \mid Y]\,dP = \frac{1}{6} \cdot P(\{\omega_1, \omega_2\}) = \frac{1}{6} \cdot 0.60 = 0.10$$


$$\int_A X\,dP = X(\omega_1)P(\{\omega_1\}) + X(\omega_2)P(\{\omega_2\}) = 0 \cdot 0.50 + 1 \cdot 0.10 = 0.10$$


3. **$A = \{\omega_3, \omega_4\}$ (where $Y=1$):**

$$\int_A E[X \mid Y]\,dP = \frac{1}{2} \cdot P(\{\omega_3, \omega_4\}) = \frac{1}{2} \cdot 0.40 = 0.20$$


$$\int_A X\,dP = X(\omega_3)P(\{\omega_3\}) + X(\omega_4)P(\{\omega_4\}) = 0 \cdot 0.20 + 1 \cdot 0.20 = 0.20$$


4. **$A = \Omega$:**

$$\int_\Omega E[X \mid Y]\,dP = 0.10 + 0.20 = 0.30$$


$$\int_\Omega X\,dP = 0.10 + 0.20 = 0.30$$



The defining property is verified across all conditions.

### 6. Find the Conditional PMF $p_{X|Y}(x \mid y)$

Using $p_{X|Y}(x \mid y) = \frac{P(X=x, Y=y)}{P(Y=y)}$:

* **Given $Y = 0$:**

$$p_{X|Y}(0 \mid 0) = \frac{0.50}{0.60} = \frac{5}{6}, \quad p_{X|Y}(1 \mid 0) = \frac{0.10}{0.60} = \frac{1}{6}$$


* **Given $Y = 1$:**

$$p_{X|Y}(0 \mid 1) = \frac{0.20}{0.40} = \frac{1}{2}, \quad p_{X|Y}(1 \mid 1) = \frac{0.20}{0.40} = \frac{1}{2}$$



### 7. Verify standard identity $\mathbb{E}[X\mid Y=y] = \sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid y)$

* **For $y = 0$:** $\sum x \cdot p_{X|Y}(x \mid 0) = (0 \cdot \frac{5}{6}) + (1 \cdot \frac{1}{6}) = \frac{1}{6}$ (Matches $E[X|Y=0]$)
* **For $y = 1$:** $\sum x \cdot p_{X|Y}(x \mid 1) = (0 \cdot \frac{1}{2}) + (1 \cdot \frac{1}{2}) = \frac{1}{2}$ (Matches $E[X|Y=1]$)

### 8. Calculate the Mean Square Error (MSE)

$$\text{MSE} = E\left[(X - E[X \mid Y])^2\right] = \sum_{\omega_i \in \Omega} (X(\omega_i) - E[X|Y](\omega_i))^2 \cdot P(\{\omega_i\})$$

* $\omega_1: (0 - \frac{1}{6})^2 \cdot 0.50 = \frac{1}{36} \cdot \frac{1}{2} = \frac{1}{72}$
* $\omega_2: (1 - \frac{1}{6})^2 \cdot 0.10 = \frac{25}{36} \cdot \frac{1}{10} = \frac{25}{360} = \frac{5}{72}$
* $\omega_3: (0 - \frac{1}{2})^2 \cdot 0.20 = \frac{1}{4} \cdot \frac{1}{5} = \frac{1}{20}$
* $\omega_4: (1 - \frac{1}{2})^2 \cdot 0.20 = \frac{1}{4} \cdot \frac{1}{5} = \frac{1}{20}$

$$\text{MSE} = \left(\frac{1}{72} + \frac{5}{72}\right) + \left(\frac{1}{20} + \frac{1}{20}\right) = \frac{6}{72} + \frac{2}{20} = \frac{1}{12} + \frac{1}{10} = \frac{5 + 6}{60} = \frac{11}{60} \approx 0.1833$$

### 9. Ignoring $Y$ and the Law of Total Variance

If the bank uses $E[X] = 0.30$ as a constant estimate, the error becomes the global variance $\text{Var}(X)$.
Since $X$ is Bernoulli with $p = 0.30$:


$$\text{MSE}_{\text{global}} = \text{Var}(X) = p(1-p) = 0.30 \cdot 0.70 = 0.21$$

**Relationship to Law of Total Variance:**
According to the Law of Total Variance:


$$\text{Var}(X) = E[\text{Var}(X \mid Y)] + \text{Var}(E[X \mid Y])$$

* $E[\text{Var}(X \mid Y)]$ is the MSE when using $Y$ ($\frac{11}{60} \approx 0.1833$).
* $\text{Var}(E[X \mid Y])$ is the variance explained by knowing $Y$.

By ignoring $Y$, the baseline error increases from $0.1833$ to $0.21$. The difference ($0.21 - 0.1833 = 0.0267$) is exactly $\text{Var}(E[X \mid Y])$, the variance structural knowledge stripped away from the estimator.

### 10. Information Theory Computations

* **a) Entropy $H(X)$:**

$$H(X) = -0.3\log_2(0.3) - 0.7\log_2(0.7) \approx 0.8813 \text{ bits}$$


* **b) Conditional Entropy $H(X \mid Y)$:**

$$H(X \mid Y) = P(Y=0)H(X \mid Y=0) + P(Y=1)H(X \mid Y=1)$$


$$H(X \mid Y=0) = -\frac{1}{6}\log_2\left(\frac{1}{6}\right) - \frac{5}{6}\log_2\left(\frac{5}{6}\right) \approx 0.6500$$


$$H(X \mid Y=1) = -\frac{1}{2}\log_2\left(\frac{1}{2}\right) - \frac{1}{2}\log_2\left(\frac{1}{2}\right) = 1.0000$$


$$H(X \mid Y) = (0.60 \cdot 0.6500) + (0.40 \cdot 1.0000) = 0.3900 + 0.4000 = 0.7900 \text{ bits}$$


* **c) Mutual Information $I(X; Y)$:**

$$I(X; Y) = H(X) - H(X \mid Y) = 0.8813 - 0.7900 = 0.0913 \text{ bits}$$


* **d) Percentage of uncertainty removed:**

$$\frac{I(X; Y)}{H(X)} = \frac{0.0913}{0.8813} \approx 10.36\%$$

## # Q2) Conditional Density of the Multivariate Gaussian

To derive $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$, we use the definitions of joint and marginal multivariate normal distributions. By definition:


$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}$$

Let's look at the quadratic expansion inside the exponential of the joint density. Define $\mathbf{e}_X = \mathbf{x} - \boldsymbol{\mu}_X$ and $\mathbf{e}_Y = \mathbf{y} - \boldsymbol{\mu}_Y$. The joint block matrix inverse can be expressed using the **Schur Complement** of $\boldsymbol{\Sigma}_{YY}$, which is $\mathbf{S}_{XX} = \boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$:

$$\boldsymbol{\Sigma}^{-1} = \begin{bmatrix} \boldsymbol{\Sigma}_{X\mid Y}^{-1} & -\boldsymbol{\Sigma}_{X\mid Y}^{-1}\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \\ -\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{X\mid Y}^{-1} & \boldsymbol{\Sigma}_{YY}^{-1} + \boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{X\mid Y}^{-1}\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \end{bmatrix}$$

Multiplying out the joint quadratic form $(\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu})$ yields:


$$\mathbf{e}_X^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} \mathbf{e}_X - 2\mathbf{e}_X^T \boldsymbol{\Sigma}_{X\mid Y}^{-1}\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\mathbf{e}_Y + \mathbf{e}_Y^T \boldsymbol{\Sigma}_{YY}^{-1}\mathbf{e}_Y + \mathbf{e}_Y^T \boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{X\mid Y}^{-1}\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\mathbf{e}_Y$$

Grouping the terms containing $\mathbf{e}_X$ allows us to complete the square:


$$\left( \mathbf{e}_X - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\mathbf{e}_Y \right)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} \left( \mathbf{e}_X - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\mathbf{e}_Y \right) + \mathbf{e}_Y^T \boldsymbol{\Sigma}_{YY}^{-1}\mathbf{e}_Y$$

Setting $\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$, the quadratic terms become:


$$(\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) + (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$$

When substituted back into the exponential formula, the joint exponent splits cleanly:


$$\exp\left(-\frac{1}{2}(\mathbf{z}-\boldsymbol{\mu})^T\boldsymbol{\Sigma}^{-1}(\mathbf{z}-\boldsymbol{\mu})\right) = \exp\left(-\frac{1}{2}(\mathbf{x}-\boldsymbol{\mu}^*)^T\boldsymbol{\Sigma}_{X\mid Y}^{-1}(\mathbf{x}-\boldsymbol{\mu}^*)\right) \cdot \exp\left(-\frac{1}{2}(\mathbf{y}-\boldsymbol{\mu}_Y)^T\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y}-\boldsymbol{\mu}_Y)\right)$$

For the denominators, using properties of block determinants:


$$|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{YY}| \cdot |\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}| = |\boldsymbol{\Sigma}_{YY}| \cdot |\boldsymbol{\Sigma}_{X\mid Y}|$$

Dividing the joint density $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})$ by the marginal density $f_{\mathbf{Y}}(\mathbf{y}) = \frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}}\exp\left(-\frac{1}{2}(\mathbf{y}-\boldsymbol{\mu}_Y)^T\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y}-\boldsymbol{\mu}_Y)\right)$ completely cancels the marginal $\mathbf{Y}$ terms, leaving exactly:

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{X\mid Y}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right)$$

## # Q3) Continuous Fault Monitoring of a Spindle Motor

### 1. Joint Distribution

Since $X$ and $Q$ are independent normal random variables, any linear combination of $X$ and $Y = X+Q$ is also normally distributed. Thus, $(X, Y)$ is jointly multivariate normal.

* **Show $Y \sim \mathscr{N}(0, 5)$:**

$$E[Y] = E[X + Q] = E[X] + E[Q] = 0 + 0 = 0$$


$$\text{Var}(Y) = \text{Var}(X + Q) = \text{Var}(X) + \text{Var}(Q) = 1 + 4 = 5 \quad \text{(since } X \perp Q\text{)}$$


* **Find $\text{Cov}(X, Y)$:**

$$\text{Cov}(X, Y) = \text{Cov}(X, X + Q) = \text{Cov}(X,X) + \text{Cov}(X,Q) = \text{Var}(X) + 0 = 1$$



### 2. Information Geometry & $\sigma(Y)$-measurability

$X$ is **not** $\sigma(Y)$-measurable because there is no deterministic function $g$ such that $X = g(Y)$. The mapping from the underlying state space to $Y$ is not injective due to the noise variable $Q$.

In practical terms, a power reading of $Y = 5$ could be caused by an overheating engine ($X=5, Q=0$), a routine cutting load on hard material ($X=0, Q=5$), or any infinite linear combinations in between (e.g., $X=1, Q=4$). Because multiple distinct physical states $(\omega_X, \omega_Q)$ map to the exact same observation $Y$, the engineer cannot mathematically reconstruct $X$ perfectly from $Y$ alone.

### 3. Conditional Expectation (The Best Estimate)

Using the property for joint normals:


$$E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y) = 0 + \frac{1}{5}(y - 0) = \frac{1}{5}y$$

**Signal-to-Total-Variance Ratio:**
This can be written explicitly as:


$$E[X \mid Y] = \left(\frac{\sigma_X^2}{\sigma_X^2 + \sigma_Q^2}\right) Y$$


Here, $\frac{1}{5}$ represents the ratio of true underlying system signal variance to the total combined system variance.

### 4. Verification of the Residual

Let $\widehat{X} = \frac{1}{5}Y$. The residual is $Z = X - \frac{1}{5}Y = X - \frac{1}{5}(X + Q) = \frac{4}{5}X - \frac{1}{5}Q$.

* **Prove $E[Z] = 0$:**

$$E[Z] = \frac{4}{5}E[X] - \frac{1}{5}E[Q] = \frac{4}{5}(0) - \frac{1}{5}(0) = 0$$


* **Verify Uncorrelated (Orthogonality Principle):**

$$\text{Cov}(Z, Y) = \text{Cov}\left(\frac{4}{5}X - \frac{1}{5}Q, X + Q\right) = \frac{4}{5}\text{Cov}(X,X) + \frac{4}{5}\text{Cov}(X,Q) - \frac{1}{5}\text{Cov}(Q,X) - \frac{1}{5}\text{Cov}(Q,Q)$$


$$\text{Cov}(Z, Y) = \frac{4}{5}(1) + 0 - 0 - \frac{1}{5}(4) = \frac{4}{5} - \frac{4}{5} = 0$$



### 5. Information Theory: Differential Entropy

The differential entropy of a normal distribution $\mathscr{N}(0, \sigma^2)$ is $h = \frac{1}{2}\ln(2\pi e \sigma^2)$.

* **Compute $h(X)$:**

$$h(X) = \frac{1}{2}\ln(2\pi e (1)) = \frac{1}{2}\ln(2\pi e)$$


* **Compute $h(X \mid Y)$:**
The conditional variance is $\sigma_{X|Y}^2 = 1 - \frac{1^2}{5} = \frac{4}{5} = 0.8$.

$$h(X \mid Y) = \frac{1}{2}\ln\left(2\pi e \left(\frac{4}{5}\right)\right) = \frac{1}{2}\ln(2\pi e) + \frac{1}{2}\ln(0.8)$$



### 6. Mutual Information

* **Calculate $I(X; Y)$:**

$$I(X; Y) = h(X) - h(X \mid Y) = \frac{1}{2}\ln(2\pi e) - \left[\frac{1}{2}\ln(2\pi e) + \frac{1}{2}\ln(0.8)\right] = -\frac{1}{2}\ln(0.8) = \frac{1}{2}\ln(1.25) \approx 0.1116 \text{ nats}$$


* **Uncertainty Percentage Removed:**
Using the variance metric directly to show reduction in structural uncertainty: the variance drops from $1$ to $0.8$, meaning **$20\%$** of the temperature uncertainty variance is removed.
* **If $\text{Var}(Q) = 100$:**
Then $\text{Var}(Y) = 101$, and $\text{Cov}(X,Y)=1$. The conditional variance becomes $\sigma_{X|Y}^2 = 1 - \frac{1}{101} = \frac{100}{101} \approx 0.9901$. The uncertainty removed would drop drastically to only **$0.99\%$**, as the signal gets completely drowned out by noise.

### 7. Diagnostic Interpretation

The manager's statement assumes that $Y$ is an uncorrupted proxy for $X$. However, because $Y = X + Q$, using $Y$ directly overestimates the magnitude of deviations because it includes the variance of $Q$.

Mathematically, $E[Y] = E[X]$, but as an *estimator for individual observations*, $Y$ has a structural noise bias. The optimal minimum-MSE estimator $E[X \mid Y] = \frac{1}{5}Y$ acts as a **shrinkage estimator**. Because the background noise ($Q$) accounts for $80\%$ of the total variance in the readings, any unexpected spike observed in $Y$ is statistically far more likely to be a temporary mechanical torque peak rather than a pure temperature spike. The math automatically "shrinks" the raw measurement back toward the known unconditional baseline mean ($0$) to avoid triggering false alarms.
